```python
#coding:utf-8

import requests as rq
import csv
import time
```

```python
# 处理 json 格式的电影数据，返回数据结果
def parseMovie(mv):
    if mv:
        mv_id = mv["id"].split("/")[-1]
        alt_title = mv["title"]
        author = mv["directors"][0]["name"]
        rating = mv["rating"]["average"]
        cast = "/".join( [i["name"] for i in mv["casts"]] )
        image = mv["images"]["small"]
        csv_line = [mv_id, alt_title, author, rating, cast, image]
        return csv_line, image

# 获取电影海报，以电影 ID 命名
def writeImageFile(movie_id, image_url):
    with open("./%s.jpg" % movie_id, 'wb') as wf:
        req = rq.get(image_url)
        wf.write(req.content)
        
```

```python
# 程序开始
res = [] 

for i in range(13):
    url = "https://api.douban.com/v2/movie/top250?start=%d&apikey=0df993c66c0c636e29ecbb5344252a4a" % (i*20)
    print ("爬取链接：%s\n" % url)
    try:
        req = rq.get(url)
        assert (len(req.json()["subjects"])>0)
    except:
        print("第 %d 页获取数据出错，请注意检查爬取结果数量\n" % i)
        continue
        
    for i in range(len(req.json()["subjects"])):
        res_line, image = parseMovie(req.json()["subjects"][i])
        res.append(res_line)
    time.sleep(2)

# 结果写入 csv 文件
with open("movies.csv", "w") as wf:
    csvf = csv.writer(wf)
    csvf.writerows(res)
    
# 获取海报
with open("./movies.csv", "r") as rf:
    csvf = csv.reader(rf)
    i = 1
    
    for row in csvf:
        print ("获取第 %d 部电影海报，电影ID:%s\n" % (i,row[0]))
        writeImageFile(row[0], row[-1])
        i += 1
        time.sleep(1)
```
