```python
import pandas as pd
```

```python
#读取 Task_4_lianjia_sale.csv 数据(阅读参考文章 2、3)，利用 groupby 函数，以【区域】字段进行 groupby， 同时结合 agg 聚合函数统计各个区域平均房价;
#将统计所得平均房价命名为【区域平均价格】列 (阅读参考文章 4、5)
lianjia_data = pd.read_csv('./Task_4_lianjia_sale.csv')
#lianjia_data.groupby(['区域']).价格.agg([('区域平均价格', 'mean')])

#并将数据从小到大按顺序排列(提示:df.sort_values(by = “ ”) (阅读参考文章 6));
#lianjia_data.groupby(['区域']).价格.agg([('区域平均价格', 'mean')]).sort_values(by = '区域平均价格')

#最后展示所得前 5 行数据 (将所得的 DataFrame 命名为 df_avg);
df_avg = lianjia_data.groupby(['区域']).价格.agg([('区域平均价格', 'mean')]).sort_values(by = '区域平均价格')
df_avg
```

```python
#利用 Pyecharts 将第(1)问所得数据作条形图(阅读参考文章 7)，图表标题为“上海各区域平均价格”(提示: 最后用 render_notebook()渲染显示出图片);
bar = (
    Bar()
    .add_xaxis(df_avg.index.tolist())
    .add_yaxis(df_avg.区域平均价格.name, df_avg.区域平均价格.tolist())
    #.set_series_opts(label_opts=opts.LabelOpts(formatter = '{c}'))
    .set_global_opts(title_opts=opts.TitleOpts(title = '上海各行政区平均价格'))
)

bar.render_notebook()
```

```python
#利用 groupby 函数，以【装修类型】字段进行 groupby，同时结合 agg 聚合函数统计各个区域房源数量;
#将统计所得数据命名为【房源数量】列;最后展示所得数据(将所得的 DataFrame 命名为 df_zhuangxiu);
df_zhuangxiu = lianjia_data.groupby(['装修类型']).房型.agg([('房源数量', 'count')])
df_zhuangxiu
```

```python
#利用 Pyecharts 将第(3)问所得数据作饼图(阅读参考文章 7、8)，图表标题为“装修类型占比情况”
#(提示: v1=["啤酒","可乐","雪碧","咖啡","奶茶"] v2=[30,19,21,12,18]，a = [list(z) for z in zip(v1,v2)]， 则生成 v1 与 v2 一一对应的数据)。
pie_data = df_zhuangxiu.sort_values(by = '房源数量', ascending = False)
pie = (
    Pie()
    .add('', [list(z) for z in zip(pie_data.房源数量.index, pie_data.房源数量.tolist())])
    .set_series_opts(label_opts=opts.LabelOpts(formatter = "{b}:{c}({d}%)"))
    .set_global_opts(title_opts=opts.TitleOpts(title = '装修类型占比情况'))
)

pie.render_notebook()
```
