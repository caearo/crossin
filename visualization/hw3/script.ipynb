```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
#读取 CSV 格式文件“task_3.1_zhihu_timeline_answer.csv”，将数据命名为 zhihu_data_answer，并查看前部数据(阅读参考文章 1、2);

zhihu_data_answer = pd.read_csv('./task_3.1_zhihu_timeline_answer.csv')
zhihu_data_answer.head(2)
```

```python
#提取“task_3.1_zhihu_timeline_answer.csv”中【回答点赞数】和【回答感谢数】列数据，制成散点图， 观察两者的关系(阅读参考文章 3、4);

dimX = zhihu_data_answer.回答点赞数
dimY = zhihu_data_answer.回答感谢数

fig = plt.figure()  
ax1 = fig.add_subplot(111)

# 设置图片大小
plt.rcParams['figure.figsize'] = (10,6)

#设置标题  
ax1.set_title('回答点赞数和回答感谢数', fontsize = 'xx-large', fontweight = 'heavy')  

#设置轴标签  
plt.xlabel(dimX.name)  
plt.ylabel(dimY.name)  

#画散点图  
ax1.scatter(dimX, dimY, c = 'c', marker = 'x', alpha = .6)  

#显示所画的图  
plt.show() 
```

```python
#提取“task_3.1_zhihu_timeline_answer.csv”中【问题回答数】和【问题关注数】列数据，制成散点图， 观察两者的关系;

dimX = zhihu_data_answer.问题回答数
dimY = zhihu_data_answer.问题关注数

plt.title('问题回答数和问题关注数', fontsize = 'xx-large', fontweight = 'heavy')

plt.xlabel(dimX.name)
plt.ylabel(dimY.name)

plt.scatter(dimX, dimY, c = 'c', marker = 'd', alpha = .4)

plt.show()
```

```python
#读取 CSV 格式文件“task_3.2_zhihu_timeline_article.csv”，将数据命名为 zhihu_data_article，并查看 前部数据;

zhihu_data_article = pd.read_csv('./task_3.2_zhihu_timeline_article.csv')
zhihu_data_article.head(2)
```

```python
#提取“task_3.2_zhihu_timeline_article.csv”中【文章点赞数】和【文章评论数】列数据，制成散点图， 观察两者的关系;

dimX = zhihu_data_article.文章点赞数
dimY = zhihu_data_article.文章评论数

plt.title('文章点赞数和文章评论数', fontsize = 'xx-large', fontweight = 'heavy')

plt.xlabel(dimX.name)
plt.ylabel(dimY.name)

plt.scatter(dimX, dimY, c = 'c', marker = 'd', alpha = .4)

plt.show()
```

```python
#提取“task_3.2_zhihu_timeline_article.csv”中【文章点赞数】低于 1000 的行数据，作出箱型图(阅读参 考文章 5)。

data = zhihu_data_article[zhihu_data_article.文章点赞数 < 1000].文章点赞数

plt.title('文章点赞数分布', fontsize = 'xx-large', fontweight = 'heavy')

plt.ylabel(data.name)

plt.boxplot(data)

plt.show()
```
